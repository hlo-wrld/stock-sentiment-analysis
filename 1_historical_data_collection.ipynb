```python
import yfinance as yf
import pandas as pd

# List of stock tickers
stock_tickers = ["AAPL", "MSFT", "GOOGL", "AMZN"]

# Create an empty DataFrame
combined_df = pd.DataFrame()

# Loop through each stock ticker
for ticker in stock_tickers:
    getinfo = yf.Ticker(ticker)
    df = getinfo.history(period="5y")
    # Add a new column with the stock ticker
    df['Ticker'] = ticker
    # Concatenate the DataFrame with the combined DataFrame
    combined_df = pd.concat([combined_df, df])

# Reset index
combined_df.reset_index(inplace=True)
(combined_df)
```

```python
combined_df['Previous Close'] = combined_df.groupby('Ticker')['Close'].shift(1)
(combined_df.head())
```

```python
combined_df['Movement'] = combined_df.apply(lambda row: 'Increase' if row['Close'] > row['Previous Close'] else 'Decrease', axis=1)
(combined_df.head())
```

```python
combined_df.dropna(subset=['Previous Close'], inplace=True)
```
