```python
import pandas as pd
df_sent= pd.read_csv('C:\\Users\\Admin\\Desktop\\stock_analysis\\sentiment_data.csv')
df_hilo= pd.read_csv('C:\\Users\\Admin\\Desktop\\stock_analysis\\high_low_data1.csv')                
```

```python
df_sent.head()
```

```python
df_hilo.head()
```

```python
df_sent['Date'] = pd.to_datetime(df_sent['Date'], format='%d-%m-%Y').dt.date.astype(str)
```

```python
df_hilo['Date'] = pd.to_datetime(df_hilo['Date'], utc=True).dt.strftime('%Y-%m-%d')
```

```python
df_hilo.head()
```

```python
df_sent.tail()
```

```python
df_sent.rename(columns={'Stock Symbol': 'Ticker'}, inplace=True)
```

```python
df_merged = pd.merge(df_hilo, df_sent, left_on=['Date', 'Ticker'], right_on=['Date', 'Ticker'], how='left')
```
